```python
import pandas as pd
data = pd.read_csv("movies.csv",encoding = 'utf_8_sig')
data.head()
```

```python
row,col = data.shape
```

```python
data['新分類'] = "create"
for i in range(0,row):
    if data['分類'][i][4:5] == "/":
        movie_type_1 = data['分類'][i][2:7]
        #print(i)
    else:
        movie_type_1 = data['分類'][i][2:4]
    #movie_type_1 = data['分類'][i][4:5]
        
    data.loc[i,'新分類'] = movie_type_1
data
```

```python
data.to_csv( 'movies.csv', encoding='utf_8_sig' , index = False)
```

```python
from sklearn.model_selection import train_test_split

train_data , test_data = train_test_split(data,test_size=0.2)
```

```python
import jieba
import re
import jieba.analyse
import re
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics



def content_cut(text):
    specialChars = "！#$%^&*（）《  》【]「"
    text = text.replace("★" , "").replace("\r\n" , "").replace("！","").replace("《","").replace("》","").replace("，","").replace("/","").replace("（","").replace("）","")
    return " ".join(list(jieba.analyse.extract_tags(text,topK=2)))

data['分詞劇情'] = data['劇情介紹'].map(content_cut)
#test_data['分詞劇情'] = test_data['劇情介紹'].map(content_cut)
```

```python
data = { 'type':data["新分類"] ,'name':data["中文名稱"],'article':data['分詞劇情']}
df1 = pd.DataFrame(data)

df1['key'] = df1['name'].astype(str)+' '+df1['type'].astype(str)+' '+df1['article'].astype(str)
```

```python
def count(df):
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(df['key'])
    tfidf = TfidfTransformer() 
    tf=tfidf.fit_transform(X)
    word = vectorizer.get_feature_names() #詞袋 
    
    return tf

def knn(X,Y):
    X_train, X_test, y_train, y_test = train_test_split(X, Y['type'].str[0:2], test_size=0.2)
    clf=KNeighborsClassifier(n_neighbors=51)
    clf.fit(X_train,y_train)

    y_pred = clf.predict(X_test)    # 預測模型 
    y_test=y_test.values

    print("Test set score:{:2f}",metrics.accuracy_score(y_test, y_pred))

tf = count(df1)
knn(tf,df1)
```
