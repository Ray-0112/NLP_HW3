```python
import pandas as pd
data = pd.read_csv("movies.csv",encoding = 'utf_8_sig')
data.head()
```

```python
row,col = data.shape
```

```python
data['新分類'] = "create"
for i in range(0,row):
    movie_type = data['分類'][i][2:4]
    data.loc[i,'新分類'] = movie_type
data
```

```python
data.to_csv( 'movies.csv', encoding='utf_8_sig' , index = False)
```

```python
from sklearn.model_selection import train_test_split

train_data , test_data = train_test_split(data, random_state = 42,test_size=0.2)
```
